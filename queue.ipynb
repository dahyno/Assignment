<h1>1. LeetCode 225</h1>

```python
from collections import deque

class MyStack:
    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.queue = deque()

    def push(self, x: int) -> None:
        """
        Push element x onto stack.
        """
        self.queue.append(x)
        # Rotate elements to maintain stack order
        for _ in range(len(self.queue) - 1):
            self.queue.append(self.queue.popleft())

    def pop(self) -> int:
        """
        Removes the element on top of the stack and returns it.
        """
        return self.queue.popleft()

    def top(self) -> int:
        """
        Get the top element without removing it.
        """
        return self.queue[0]

    def empty(self) -> bool:
        """
        Returns whether the stack is empty.
        """
        return not len(self.queue)

# 사용 예시:
stack = MyStack()
stack.push(1)       # queue: [1]
stack.push(2)       # queue: [2,1] (회전 후)
stack.push(3)       # queue: [3,2,1] (회전 후)
print(stack.pop())  # 3
print(stack.top())  # 2
print(stack.empty())# False
```

<h1>2. LeetCode 232</h1>

```python
class MyQueue:
    def __init__(self):
        self.input = []
        self.output = []

    def push(self, x: int) -> None:
        self.input.append(x)

    def pop(self) -> int:
        self.peek()  # Ensure output has the front element
        return self.output.pop()

    def peek(self) -> int:
        if not self.output:
            # Transfer all elements from input to output (reverse order)
            while self.input:
                self.output.append(self.input.pop())
        return self.output[-1]

    def empty(self) -> bool:
        return not self.input and not self.output

# 사용 예시:
obj = MyQueue()      # MyQueue 객체 생성
obj.push(1)          # 큐에 1 추가
obj.push(2)          # 큐에 2 추가
print(obj.peek())    # 출력: 1 (맨 앞 요소 확인)
print(obj.pop())     # 출력: 1 (맨 앞 요소 제거 및 반환)
print(obj.empty())   # 출력: False (큐가 비어 있지 않음)
```
